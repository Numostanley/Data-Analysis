```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn import datasets
```

```python
from sklearn.cluster import KMeans
iris = datasets.load_iris()
flow = pd.DataFrame(iris.data)
flow1 = pd.DataFrame(zip(flow[0],flow[1]))
flow1.columns = ["x", "y"]
kmeans = KMeans(n_clusters=3)
kmeans.fit(flow1)
```

```python
label = kmeans.predict(flow1)
centroids = kmeans.cluster_centers_
```

```python
fig = plt.figure(figsize=(5,5))

k = 3
colmap = {(x+1): [(np.sin(0.3*x + 0)*127+128)/255,(np.sin(0.3*x + 2)*127+128)/255,(np.sin(0.3*x + 4)*127+128)/255] for x in range(k)} # making rainbow colormap
colors = map(lambda x: colmap[x+1], label)
colors1 = list(colors)
plt.scatter(flow1["x"], flow1["y"], color= colors1, alpha=0.5, edgecolor = 'k')
for idx, centroids in enumerate(centroids):
    plt.scatter(*centroids, color=colmap[idx+1])
plt.xlim(0,80)
plt.ylim(0,80)
plt.show()
```
